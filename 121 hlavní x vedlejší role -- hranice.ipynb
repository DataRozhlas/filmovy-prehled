Kde leží hranice mezi hlavní a vedlejší rolí v závislosti na pořadí v seznamu "Hrají"? Pomocníkem nám budou nominace na České lvy.

```python
import pandas as pd
```

```python
filmy = pd.read_json("filmy.json")
```

```python
filmy = filmy[["Film","Hrají"]]
```

```python
filmy = filmy.explode("Hrají")
```

```python
filmy['idx'] = filmy.groupby(level=0).cumcount()
```

```python
filmy[filmy["Film"] == "Je třeba zabít Sekala (163291)"]
```

```python
lvi = pd.read_json("lvi_herectvo.json")
```

```python
lvi
```

```python
lvi = lvi.explode("Hlavní role").explode("Vedlejší role")
```

```python
hv = filmy.merge(lvi, on="Film")
hv.sample(60)
```

```python
vedlejsi_role = hv[hv["Hrají"] == hv["Vedlejší role"]]
```

```python
vedlejsi_role = vedlejsi_role.groupby(["Film","Vedlejší role"]).first()
```

```python
vedlejsi_role
```

```python
vedlejsi_role.idx.max()
```

```python
vedlejsi_role.idx.mean()
```

```python
hlavni_role = hv[hv["Hrají"] == hv["Hlavní role"]]
```

```python
hlavni_role = hlavni_role.groupby(["Film","Hlavní role"]).first()
```

```python
hlavni_role
```

```python
hlavni_role.idx.mean()
```
