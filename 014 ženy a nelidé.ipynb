## Nelidé

```python
df = pd.read_json(os.path.join("data", "filmy.json"))
```

```python
hraji = df.explode("Hrají")["Hrají"].dropna()
```

```python
mask = hraji.str[0].str.islower()
```

```python
result = hraji[mask]
```

```python
result.drop_duplicates().to_list()
```

```python
nejsou_lide = result.drop_duplicates().to_list()
```

```python
persony = pd.read_json(os.path.join("data", "persony.json"))
```

```python
persony = persony[persony["Osoba"] !="fyzická"]
```

```python
firmy = persony["Jméno"].drop_duplicates().to_list()
firmy
```

```python
for f in firmy:
    nejsou_lide.append(f)
```

```python
with open(os.path.join("data_fixes", "nelide.json"), "w+", encoding="utf-8") as nelide:
    json.dump(nejsou_lide, nelide)
```

## Ženy

```python
df_raw = pd.read_json(os.path.join("data", "filmy_raw.json"))
```

```python
df["Vedoucí produkce"].explode().drop_duplicates().to_list()
```

```python
df["Vedoucí produkce"].drop_duplicates().to_list()
```

```python
vsechna_jmena = []
for p in profese:
    pracovni = df_raw[p]
    seznam_jmen = pracovni.explode().drop_duplicates().to_list()
    for s in seznam_jmen:
        if s not in vsechna_jmena:
            vsechna_jmena.append(s)
zeny = []
for v in vsechna_jmena:
    if v:
        if isinstance(v, list) or isinstance(v, str):
            v = v.split(",")
            if isinstance(v, list):
                for w in v:
                    if "/ž" in w:
                        if w not in zeny:
                            zeny.append(w.strip())
            if isinstance(v, str):
                if "/ž" in v:
                    if v not in zeny:
                        zeny.append(v.strip)
```

```python
zeny
```

```python
zeny_ciste = [x.split(" /ž/")[0].replace("české verze ","") for x in zeny]
zeny_ciste
```

```python
with open(os.path.join("data_fixes", "zeny.json"), "w+", encoding="utf-8") as zeny_soubor:
    json.dump(zeny_ciste, zeny_soubor)
```
