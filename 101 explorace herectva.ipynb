## Bohdalka

```python
hh = herectvo[herectvo["Úloha"] == "Hrají"]
```

```python
debut = pd.Series(hh.groupby("Jméno")["Rok"].min(), name="Debut")
hh = hh.merge(debut.to_frame(), on="Jméno")
hh["Debut"] = pd.to_numeric(hh["Debut"], errors='coerce')
```

```python
derniera = pd.Series(hh.groupby("Jméno")["Rok"].max(), name="Derniéra")
hh = hh.merge(derniera.to_frame(), on="Jméno")
hh["Derniéra"] = pd.to_numeric(hh["Derniéra"], errors='coerce')
```

```python
hh["Kariéra"] = hh["Derniéra"] - hh["Debut"]
```

```python
hh.groupby("Jméno")["Kariéra"].max().nlargest(10)
```

```python
hh[hh["Věk"] < 7].groupby("Jméno")["Kariéra"].max().nlargest(10)
```

```python
hh[hh["Jméno"] == "Jiřina Bohdalová (2878)"]
```

```python
hh["spoluherectvo"] = hh["Titul"].map(hh.groupby("Titul")["Jméno"].agg(list))
```

```python
bohdalka = hh[hh["Jméno"] == "Jiřina Bohdalová (2878)"]["spoluherectvo"].to_list()
bohdalka2 = [item for bohdalka in bohdalka for item in bohdalka]
bohdalka = []
for b in bohdalka2:
    if b not in bohdalka:
        bohdalka.append(b)
print(len(bohdalka))
```

```python
hh[hh.Jméno.isin(bohdalka)].groupby("Jméno")["Debut"].min().nsmallest()
```

```python
def kolegove(jmeno):
    hh["spoluherectvo"] = hh["Film"].map(hh.groupby("Film")["Jméno"].agg(list))
    kolegove = hh[hh["Jméno"] == jmeno]["spoluherectvo"].to_list()
    kolegove2 = [item for kolegove in kolegove for item in kolegove]
    kolegove = []
    for k in kolegove2:
        if k not in kolegove:
            kolegove.append(k)
    return(kolegove)
```

```python
def gender(jmeno):
    gender = ""
    jmeno = jmeno.split("(")[0].strip()
    if jmeno[-1:] == "á":
        gender = "žena"
    elif jmeno[-3:] == "ova":
        gender = "žena"
    elif "Vica" in jmeno:
        gender = "žena"
    elif "Chantal" in jmeno:
        gender = "žena"
    elif "Lilian" in jmeno:
        gender = "žena"
    elif "Nataša" in jmeno:
        gender = "žena"
    elif "Beata" in jmeno:
        gender = "žena"
    else:
        gender = "muž"
    return gender
```

```python
for x in tisic:
    print(f"{x} {gender(x)}")
```

```python
from wikidata.client import Client
import wikidata.entity
```

```python
def wikidates(id):
    # zřejmě absurdně neefektivní kód, ale potřebuju ho jen jednou a z GPT-4 to fungovalo
    client = Client()  # doctest: +SKIP
    entity = client.get(id, load=True)
    P569_entity = wikidata.entity.Entity("P569", client)
    entity_dict = dict(entity)
    date_of_birth = entity_dict[P569_entity]
    try:
        P570_entity = wikidata.entity.Entity("P570", client)
        date_of_death = entity_dict[P570_entity]
        print(date_of_death)
    except KeyError:
        date_of_death = None
    
    return [pd.to_datetime(date_of_birth), pd.to_datetime(date_of_death)]
```
