```python
from bs4 import BeautifulSoup
import requests
import os
```

```python
def get_domain(http): 
    domain = "https://" + http.split("/")[2]
    return(domain)
```

```python
def get_all_links(url, printurl=True):
    domain = get_domain(url)
    if printurl:
        print(url)
    urls_on_page = []
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    for link in soup.find_all('a'):
        link_url = link.get('href')
        if link_url not in urls_on_page:
            try:
                if link_url[0] == "/":
                    link_url = domain + link_url
                if domain in link_url:
                    urls_on_page.append(link_url)
            except:
                pass
    return(urls_on_page)
```

```python
def clean_list(source_string):
    cleaned_list = source_string.splitlines()
    cleaned_list = list(filter(lambda x: x.strip(), cleaned_list))
    cleaned_list = list(set(cleaned_list))
    return(cleaned_list)
```

```python
from datetime import datetime
```

```python
def clean_list(file):
    file.seek(0)  # Seek to the beginning of the file
    contents = file.read().splitlines()
    return [line.strip() for line in contents if line.strip()]

def crawl_web(url):

    crawled_domain = get_domain(url).replace("https://", "")

    filename_all = crawled_domain + "_all" + ".txt"
    filename_visited = crawled_domain + "_visited" + ".txt"
    filename_unvisited = crawled_domain + "_unvisited" + ".txt"

    with open(os.path.join("crawl", filename_all), "a+", encoding="utf-8") as file_all:
        urls_all = clean_list(file_all)
    with open(os.path.join("crawl", filename_visited), "a+", encoding="utf-8") as file_visited:
        urls_visited = clean_list(file_visited)
    with open(os.path.join("crawl", filename_unvisited), "a+", encoding="utf-8") as file_unvisited:
        urls_unvisited = clean_list(file_unvisited)

    urls_unvisited = list(set(urls_unvisited) - set(urls_visited))

    if url not in urls_unvisited:
        urls_unvisited.append(url)

    urls_all = list(set(urls_all + urls_visited + urls_unvisited))

    while urls_unvisited:
        unvisited = urls_unvisited.pop(0)
        new_links = get_all_links(unvisited, printurl=True)
        urls_visited.append(unvisited)

        for new_link in new_links:
            new_link = new_link.split("#")[0]
            if new_link not in urls_visited and new_link not in urls_unvisited:
                urls_unvisited.append(new_link)
                urls_all.append(new_link)

        with open(os.path.join("crawl", filename_all), "w", encoding="utf-8") as file_all:
            file_all.write("\n".join(urls_all))
        with open(os.path.join("crawl", filename_visited), "w", encoding="utf-8") as file_visited:
            file_visited.write("\n".join(urls_visited))
        with open(os.path.join("crawl", filename_unvisited), "w", encoding="utf-8") as file_unvisited:
            file_unvisited.write("\n".join(urls_unvisited))

        no_visited = len(urls_visited)
        no_unvisited = len(urls_unvisited)
        no_all = len(urls_all)
        no_control = no_visited + no_unvisited - no_all
        t = datetime.now().strftime("%H:%M:%S")
        print(f"{t} / visited: {no_visited} / univisited: {no_unvisited} / total: {no_all} / control: {no_control}")

    return urls_all
```

```python
crawl_web("https://www.filmovyprehled.cz/")
```
