```python
import os
import pandas as pd
```

```python
import simplemma
```

```python
lvi = pd.read_json(os.path.join('data','ceske_lvy.json'))
lvi.loc[lvi['Film'] == 'Záhrada', 'Film'] = 'Zahrada (8967)'
filmy = pd.read_json(os.path.join('data','filmy.json'))
filmy = filmy[filmy['Copyright'] > 1993]
typologie = filmy.explode("Typologie")
celoveceraky = typologie[(typologie['Typologie'] == 'hraný') & (typologie['Délka'] == 'dlouhometrážní')]
celoveceraky = celoveceraky['Film'].drop_duplicates().to_list()
filmy = filmy[filmy['Film'].isin(celoveceraky)]
```

```python
nejlepsifilmy = lvi[(lvi['KategorieS'] == 'Nejlepší film') & (lvi['Lev'] == 'vyhra')]['Film'].drop_duplicates().to_list()
nominovanefilmy = lvi[(lvi['KategorieS'] == 'Nejlepší film')]['Film'].drop_duplicates().to_list()
```

```python
def klicovaslova(obsah):
    
    def remove_non_letters(input_string):
        # Define an empty string to store our result
        result_string = ''

        # Loop through each character in the given string
        for character in input_string:
            # If the character is a letter or whitespace
            if character.isalpha() or character.isspace():
                # Include it in our result string
                result_string += character

        # Return the cleaned string
        return result_string
    
    try:
        print(obsah[0:160])
        obsah = remove_non_letters(obsah)
        obsah = obsah.split()
        klicova = []
        for o in obsah:
            slovo = simplemma.lemmatize(o, lang='cs')
            klicova.append(slovo)
        return list(set(klicova))
    except:
        return None
```

```python
filmy['Klíčová slova'] = filmy['Obsah'].apply(lambda x: klicovaslova(x))
```

```python
klicovaslova = filmy[['Film','Copyright','Klíčová slova']].explode('Klíčová slova')
xfilmu = 20
asponxfilmu = pd.Series(klicovaslova.groupby('Klíčová slova')['Film'].nunique())
asponxfilmu = asponxfilmu[asponxfilmu >= xfilmu].index.to_list()
klicovaslova = klicovaslova[klicovaslova['Klíčová slova'].isin(asponxfilmu)]
klicovaslova['Klíčová slova'] = klicovaslova['Klíčová slova'].apply(lambda x: x.lower())
```

```python
pomerlvinelvi = (klicovaslova[klicovaslova['Film'].isin(nejlepsifilmy)].groupby("Klíčová slova")['Film'].nunique() / klicovaslova.groupby("Klíčová slova")['Film'].nunique()).dropna().sort_values()
```
