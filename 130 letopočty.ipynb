```python
import os
import pandas as pd
pd.set_option('display.max_columns', None)
df = pd.read_json(os.path.join("data", "filmy.json"))
```

```python
df.columns.to_list()
```

```python
df[df["Konec výroby"].isnull()]
```

```python
df["Rozdíl mezi copyrightem a koncem výroby"] = df["Copyright"] - df["Konec výroby"]
```

```python
df[["Film","Copyright","Konec výroby","Rozdíl mezi copyrightem a koncem výroby"]].sort_values(by="Rozdíl mezi copyrightem a koncem výroby",ascending=False).head(50)
```
