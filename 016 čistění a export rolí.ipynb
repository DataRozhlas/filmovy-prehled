## Nahrazení ids jmény

```python
def replace_names(x):
    if isinstance(x, str):
        if x.strip() in ids_jmena:
            return ids_jmena[str(x)]
        else:
            return x
    elif isinstance(x, list):
        return [str(replace_names(xi)) for xi in x]
    else:
        return x
```

```python
persony = pd.read_json(os.path.join("data", "persony.json"))
```

```python
persony = persony[["Id","Jméno"]]
```

```python
ids_jmena = {}
for index, row in persony.iterrows():
    ids_jmena[str(row["Id"])] = str(row["Jméno"])
ids_jmena
```

```python
herectvo["Hraje"] = herectvo["Hraje"].apply(replace_names)
herectvo["Mluví"] = herectvo["Mluví"].apply(replace_names)
```
