## Lidé z Hradiště

```python
persony = pd.read_json(os.path.join("data", "persony.json"))
```

```python
profese = seznamy["profese"]
```

```python
def plose_jednoduse(x):
    seznam = []
    for y in x:
        if y == None:
            pass
        if isinstance(y, str):
            if y not in seznam:
                seznam.append(y)
        if isinstance(y, list):
            for z in y:
                if z not in seznam:
                    seznam.append(z)
    return(seznam)
```

```python
df["Lidé"] = df[profese].values.tolist()
df["Lidé"] = df["Lidé"].apply(lambda x: plose_jednoduse(x))
```

```python
lide = df.explode("Lidé")
```

```python
lide = lide.merge(persony, left_on="Lidé", right_on="Jméno")
```

### Hezký přehled filmařstva z UH

```python
for l, c in lide[lide['Jméno'].isin(hradiststi)].groupby("Jméno")['Film'].nunique().nlargest(20).items():
    radek = persony[persony['Jméno'] == l]
    try:
        smrt = int(radek['Rok úmrtí'].iloc[0])
    except:
        smrt = ""
    filmy = lide[lide['Jméno'] == l].sort_values(by='Rok výroby')['Film'].apply(lambda x: x.split(' (')[0]).drop_duplicates().to_list()
    print(f"""{l.split(' (')[0].upper()} | {int(radek['Rok narození'])}-{smrt}
{radek['Medailon'].to_list()[0]}
{len(filmy)} filmů: {', '.join(filmy)} 
    """)
```
