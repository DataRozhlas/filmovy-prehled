```python
import os
import pandas as pd
df = pd.read_json(os.path.join("data", "filmy.json"))
```

```python
df = df.explode("Hrají").explode("Žánr")
```

```python
obsazovanost = pd.Series(df.groupby("Hrají")["Film"].nunique(), name="Obsazovanost")
```

```python
df = df.merge(obsazovanost, on="Hrají")
```

```python
df.shape
```

```python
", ".join(df["Žánr"].drop_duplicates().to_list())
```

```python
zanry = df["Žánr"].drop_duplicates().to_list()
```

```python
df.groupby("Žánr")["Film"].nunique().nlargest(15)
```

```python
df.sample(5)
```

```python
pd.Series(df[(df["Žánr"].str.lower().str.contains("komedie")) & (df["Obsazovanost"] >= 10)].groupby("Hrají")["Film"].nunique() / df.groupby("Hrají")["Obsazovanost"].max()).nlargest(10)
```

```python
pd.Series(df[(df["Žánr"].str.lower().str.contains("komedie")) & (df["Obsazovanost"] >= 10)].groupby("Hrají")["Film"].nunique() / df.groupby("Hrají")["Obsazovanost"].max()).nsmallest(10)
```

Obsazovanější 1/2 herectva: žebříčky podle žánrů. (Jde o poměr mezi hraním ve filmech žánru x ostatních žánrech.)

for z in zanry:
    print(z)
    print(pd.Series(df[(df["Žánr"] == z) & (df["Obsazovanost"] >= df["Obsazovanost"].median())].groupby("Hrají")["Film"].nunique() / df.groupby("Hrají")["Obsazovanost"].max()).nlargest(3))
    print("***")

Obsazovanější 1/3 herectva: žebříčky podle žánrů. (Jde o poměr mezi hraním ve filmech žánru x ostatních žánrech.)

```python
for z in zanry:
    print(z)
    print(pd.Series(df[(df["Žánr"] == z) & (df["Obsazovanost"] >= df["Obsazovanost"].quantile(0.666))].groupby("Hrají")["Film"].nunique() / df.groupby("Hrají")["Obsazovanost"].max()).nlargest(3))
    print("***")
```
