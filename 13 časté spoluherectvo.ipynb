```python
from itertools import combinations

def apriori(lst, n):
    freq = {}
    # find frequent itemsets of size one
    for l in lst:
        for item in l:
            if item in freq:
                freq[item] += 1
            else:
                freq[item] = 1

    print(freq)
                
    # Remove infrequent items
    freq = {k:v for k, v in freq.items() if v >= n}
    
    if not freq:
        return []

    #find frequent itemsets of size > 1
    lsets = [set([x]) for x in freq]
    k = 2
    
    print(lsets)
    
    while True:
        csets = frozenset([frozenset(x) for x in combinations(freq, k)])
        freq = {}

        for l in lst:
            for c in csets:
                if c.issubset(l):
                    if tuple(c) in freq:
                        freq[tuple(c)] += 1
                    else:
                        freq[tuple(c)] = 1

        freq = {k:v for k, v in freq.items() if v >= n}
        if not freq:
            break
        lsets = [set(x) for x in freq.keys()]
        k += 1

    # convert lsets to lists and return n-sized itemsets
    lsets = [list(x) for x in lsets if len(x) == n]
    return lsets
```
