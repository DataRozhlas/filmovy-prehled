```python
import requests
import re
from bs4 import BeautifulSoup 
```

```python
url = "https://cs.wikipedia.org/wiki/%C4%8Cesk%C3%BD_lev_1994"

# Send an HTTP GET request to the URL
response = requests.get(url)

# Create a BeautifulSoup object with the HTML content
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
rdk = [1994, "Nejlepší režie", "Vítěz", "Díky za každé nové ráno", "Milan Šteindler"]
```

```python
# Initialize an empty list to store the results
rocnik = []

# Find all 'h3' tags
kateg = soup.find_all(attrs={'data-groupnames': '0'})

# Loop through each 'h3' tag
for h3 in kateg:
    # Initialize a temporary list to store the items for each section
    section_list = []
    
    # Find the next sibling 'ul' tag after the current 'h3' tag
    ul = h3.find_next_sibling('ul')
    
    # Find all 'li' tags within the 'ul' tag
    li_tags = ul.find_all('li')
    
    # Append the 'h3' tag's text to the section list
    section_list.append(h3.get_text().split("[")[0])
    
    # Loop through each 'li' tag and append its text to the section list
    for li in li_tags:
        section_list.append(li.get_text())
    
    # Append the section list to the final result
    rocnik.append(section_list)

# Print the final result
print(rocnik)
```
