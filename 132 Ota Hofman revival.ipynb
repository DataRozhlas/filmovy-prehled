```python
with open(os.path.join("private", "apikey.txt")) as ak:
    ai.api_key = ak.read()
```

```python
def shrn_obsah(obsah):
    answer = ai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "Jsi pracovnice filmového archivu. Následuje popis děje filmu. Shrň ho co nejstručněji, jednou až třemi větami, ale tak, aby šlo z tohoto stručného shrnutí co nejpřesněji vyvodit, co se ve filmu dělo a jaké emoce to vyvolávalo."},
            {
                "role": "user",
                "content": f"""{obsah}""",
            },
        ],
    )
    return answer.choices[0].message.content.strip()
```

```python
def shrn_shrnuti(strucne_namety):
    answer = ai.ChatCompletion.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "Jsi filmová scénáristka a vědkyně. Následují shrnutí filmů natočených podle scénáře pana X. Shrň autorský rukopis pana X v deseti bodech tak, aby z tohoto shrnutí šlo co nejlépe rekonstruovat jeho autorský rukopis, tedy napsat scénář či námět v jeho autorském stylu."},
            {
                "role": "user",
                "content": f"""{strucne_namety}""",
            },
        ],
    )
    return answer.choices[0].message.content.strip()
```

```python
def pojmenuj_film(namet):
        answer = ai.ChatCompletion.create(
        model="gpt-4",
        messages=[
            {"role": "system", "content": "Jsi filmová producentka."},
            {
            "role": "user",
            "content": f"""Tvým úkolem je předložit 5 variant názvu konkrétního filmu. Název musí být jednoduchý, dobře zapamatovatelný, stručný, ideálně i vtipný. Musí dobře odpovídat tomuto námětu: {namet}""",
            },
        ],
    )
        return answer.choices[0].message.content.strip()
```
