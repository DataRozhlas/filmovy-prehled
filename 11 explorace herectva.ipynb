```python
import pandas as pd
```

```python
herectvo = pd.read_json("herectvo.json")
```

```python
herectvo = herectvo.explode("Role")
```

## Basic explorace

```python
herectvo.groupby("Role")["Titul"].nunique().nlargest(10)
```

```python
herectvo[herectvo.Role.str.lower().str.contains("princ", na=False)]
```

```python
herectvo[herectvo.Role.str.lower().str.contains("princ ", na=False)].sort_values(by="Věk", ascending=False)[["Jméno","Rok","Titul","Role","Věk"]].head(20)
```

```python
debut = pd.Series(herectvo.groupby("Jméno")["Rok"].min(), name="Debut")
herectvo = herectvo.merge(debut.to_frame(), on="Jméno")
herectvo["Debut"] = pd.to_numeric(herectvo["Debut"], errors='coerce')
```

## Bohdalka

```python
herectvo["spoluherectvo"] = herectvo["Titul"].map(herectvo.groupby("Titul")["Jméno"].agg(list))
```

```python
bohdalka = herectvo[herectvo["Jméno"] == "Jiřina Bohdalová"]["spoluherectvo"].to_list()
bohdalka2 = [item for bohdalka in bohdalka for item in bohdalka]
bohdalka = []
for b in bohdalka2:
    if b not in bohdalka:
        bohdalka.append(b)
print(len(bohdalka))
```

```python
herectvo[herectvo.Jméno.isin(bohdalka)].groupby("Jméno")["Debut"].min().nsmallest()
```
