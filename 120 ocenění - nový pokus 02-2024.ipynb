```python
persony = pd.read_json(os.path.join("data","persony_oceneni.json"))
```

```python
persony = persony.dropna().explode("Ocenění")
```

```python
persony.groupby("Jméno").size().nlargest(10)
```

```python
lvi = persony[persony['Ocenění'].str.lower().str.contains("český lev",na=False) & persony['Ocenění'].str.lower().str.contains("vítěz",na=False)]
```

```python
lvi
```

```python
lvi.groupby("Jméno").size().nlargest(10)
```

```python
lvi[lvi['Ocenění'].str.lower().str.contains("nejlepší režie",na=False)].groupby("Jméno").size().nlargest(10)
```

```python
lvi[lvi['Ocenění'].str.lower().str.contains("nejlepší mužsk",na=False)].groupby("Jméno").size().nlargest(10)
```

```python
lvi[lvi['Ocenění'].str.lower().str.contains("nejlepší žensk",na=False)].groupby("Jméno").size().nlargest(10)
```

```python
lvi_herectvo = lvi[lvi['Ocenění'].str.lower().str.contains("nejlepší žensk",na=False)]['Jméno'].drop_duplicates().to_list() + lvi[lvi['Ocenění'].str.lower().str.contains("nejlepší mužsk",na=False)]['Jméno'].drop_duplicates().to_list()
```

```python
lvi_herectvo
```

```python
hraji = df.explode("Hrají")
```

```python
hraji_lvi = hraji[hraji['Hrají'].isin(lvi_herectvo)]
```

```python
hraji_lvi = hraji_lvi[['Film','Hrají']].drop_duplicates()
```

```python
hraji_lvi.groupby("Film").size().nlargest(10)
```
