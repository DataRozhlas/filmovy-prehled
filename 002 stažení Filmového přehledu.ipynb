```python
def loadlist(filename):
    with open(filename, 'r') as f:
        lines = []
        for line in f:
            lines.append(line[:-1])
    return(lines)
```

```python
def filename(url):
    return("-".join(url.split("/")[-2:]) + ".html")
```

```python
def seznam_souboru(slozka):
    seznam = []
    for file in os.listdir(slozka):
        seznam.append(file)
    return(seznam)
```

```python
import os
import requests
```

```python
import datetime
```

```python
def download_site(folder, urls, url_filter):
    
    print(f"date of the run: {datetime.date.today()}")
    
    if not os.path.exists(folder):
        os.mkdir(folder)
    else:
        pass
    
    filtered_urls = [url for url in urls if url_filter in url]
    
    remaining = len(filtered_urls)
    
    for url in filtered_urls:
        
        remaining = remaining - 1
        
        print(f"downloading {url}, {remaining} remaining")
        
        if filename(url) not in seznam_souboru(folder):
        
            r = requests.get(url)

            with open(os.path.join(folder, filename(url)), 'w+', encoding='utf-8') as f:
                f.write(r.text)        
```

```python
download_site("filmy", linky, "/cs/film/")
```
