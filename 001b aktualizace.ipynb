```python
for o in od2021:
    print(o)
    try:
        response = requests.get(o)
        soup = BeautifulSoup(response.text, 'html.parser')
        for link in soup.find_all('a'):
            link_url = link.get('href')
            if ("/film/" in link_url) or ("/person/" in link_url):
                link_url = "https://www.filmovyprehled.cz" + link_url
                od2021b.append(link_url)
                print("x " + link_url)
    except:
        print("CHYBA " + o)
```
