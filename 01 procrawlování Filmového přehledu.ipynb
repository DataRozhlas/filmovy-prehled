```python
from bs4 import BeautifulSoup
import requests
```

```python
def get_domain(http): 
    domain = "https://" + http.split("/")[2]
    return(domain)
```

```python
def get_all_links(url, printurl=True):
    domain = get_domain(url)
    if printurl:
        print(url)
    urls_on_page = []
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    for link in soup.find_all('a'):
        link_url = link.get('href')
        if link_url not in urls_on_page:
            try:
                if link_url[0] == "/":
                    link_url = domain + link_url
                if domain in link_url:
                    urls_on_page.append(link_url)
            except:
                pass
    return(urls_on_page)
```

```python
import os
```

```python
def clean_list(source_file):
    cleaned_list = source_file.read().splitlines()
    cleaned_list = list(filter(lambda x: x.strip(), cleaned_list))
    cleaned_list = list(set(cleaned_list))
    return(cleaned_list)
```

```python
def crawl_web(url):
    
    from datetime import datetime
    
    crawled_domain = get_domain(url).replace("https://","")

    filename_all = crawled_domain + "_all" + ".txt"
    filename_visited = crawled_domain + "_visited" + ".txt"
    filename_unvisited = crawled_domain + "_unvisited" + ".txt"

    with open(filename_all, "r", encoding="utf-8") as file_all:
        urls_all = clean_list(file_all)
    with open(filename_visited, "r", encoding="utf-8") as file_visited:
        urls_visited = clean_list(file_visited)
    with open(filename_unvisited, "r", encoding="utf-8") as file_unvisited:
        urls_unvisited = clean_list(file_unvisited)

    for u in urls_unvisited:
        if u in urls_visited:
            urls_unvisited.remove(u)

    if url not in urls_unvisited:
        urls_unvisited.append(url)
        
    urls_all = urls_all + urls_visited + urls_unvisited
    urls_all = list(set(urls_all))

    while urls_unvisited:
        for unvisited in urls_unvisited:
            new_links = get_all_links(unvisited, printurl=True)
            urls_unvisited.remove(unvisited)
            urls_visited.append(unvisited)
            for new_link in new_links:
                if "#" in new_link:
                    new_link = new_link.split("#")[0]
                if new_link in urls_visited:
                    pass
                else:
                    if new_link not in urls_unvisited:
                        urls_unvisited.append(new_link)
                if new_link not in urls_all:
                    urls_all.append(new_link)
            with open(filename_all, "w+", encoding="utf-8") as file_all:
                file_all.write("\n".join(urls_all))
            with open(filename_visited, "w+", encoding="utf-8") as file_visited:
                file_visited.write("\n".join(urls_visited))
            with open(filename_unvisited, "w+", encoding="utf-8") as file_unvisited:
                file_unvisited.write("\n".join(urls_unvisited))

            no_visited = len(urls_visited)
            no_unvisited = len(urls_unvisited)
            no_all = len(urls_all)
            no_control = no_visited + no_unvisited - no_all
            t = datetime.now().strftime("%H:%M:%S")
            print(f"{t} / visited: {no_visited} / univisited: {no_unvisited} / total: {no_all} / control: {no_control}")

    return(urls_all)
```

```python
crawl_web("https://www.filmovyprehled.cz/")
```
