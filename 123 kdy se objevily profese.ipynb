```python
import os
import pandas as pd
pd.set_option('display.max_columns', None)
df = pd.read_json(os.path.join("data", "filmy.json"))
```

```python
for col in df.columns:
    # find the first non-NaN value in the column
    first_non_nan = df[col].dropna().index[0]
    # print out the column name and first non-NaN date
    print(f"{col}: {df.loc[first_non_nan, 'Copyright']}, {df.loc[first_non_nan, 'Originální název']}")
```
