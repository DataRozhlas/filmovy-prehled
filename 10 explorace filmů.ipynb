## Typologie

```python
def popularita(sloupec, filtr, co):
    if co == "n":
        df[df[sloupec] == filtr].groupby("Film")[["Copyright","Film"]].first().groupby("Copyright")["Film"].nunique().plot.bar(figsize=(20,3))
    if co == "min":
        df[df[sloupec] == filtr].groupby("Film")[["Copyright","Minutáž"]].first().groupby("Copyright")["Minutáž"].sum().plot.bar(figsize=(20,3))
```
