```python
import os
import pandas as pd
df = pd.read_json(os.path.join("data", "filmy.json"))
```

```python
import json
with open(os.path.join("data_fixes", "seznamy.json"), encoding="utf-8") as seznamy:
    seznamy = json.loads(seznamy.read())
```

```python
profese = seznamy["profese"]
```

```python
vsechna_jmena = []
for p in profese:
    pracovni = df[p]
    seznam_jmen = pracovni.explode().drop_duplicates().to_list()
    for s in seznam_jmen:
        if s not in vsechna_jmena:
            vsechna_jmena.append(s)
```

```python
len(vsechna_jmena)
```

```python
lide = {}
for v in vsechna_jmena:
    if v:
        if isinstance(v,str):
            if "(" not in v:
                pass
            else:
                v = v.replace(")","")
                lide[v.split("(")[0].strip()] = v.split("(")[1]
```

```python
lide
```

```python
lide = pd.DataFrame.from_dict(lide,orient="index",columns=["Id"])
```

```python
lide["Člověk"] = lide.index
```
