```python
import os
import json
import pandas as pd
from src.gender import gender
pd.set_option('display.max_columns', None)
df = pd.read_json(os.path.join("data", "filmy.json"))
persony = pd.read_json(os.path.join("data", "persony_raw.json"))
```

```python
def nejcastejsi_neodatovani(x,co):
    df_pracovni = df[["Film",x]].explode(x)
    df_pracovni = df_pracovni.merge(persony, left_on = x, right_on = "Jméno")
    df_pracovni = df_pracovni[df_pracovni[co].isnull()]
    return df_pracovni.groupby(x)["Film"].nunique().nlargest(25)
```

```python
with open(os.path.join("data_fixes", "seznamy.json"), encoding="utf-8") as seznamy:
    seznamy = json.loads(seznamy.read())
```

```python
for p in seznamy["profese"]:
    print(nejcastejsi_neodatovani(p,"Datum narození"))
```

## Doplnění dat z Wikidat

```python
from wikidata.client import Client
import wikidata.entity
```

```python
with open(os.path.join("data_fixes", "wikidata.json"), encoding="utf-8") as wikid:
    wikid = json.loads(wikid.read())
```

```python
def wikidates(id):
    # zřejmě absurdně neefektivní kód, ale potřebuju ho jen jednou a z GPT-4 to fungovalo
    client = Client()  # doctest: +SKIP
    entity = client.get(id, load=True)
    P569_entity = wikidata.entity.Entity("P569", client)
    entity_dict = dict(entity)
    date_of_birth = entity_dict[P569_entity].strftime('%d.%m.%Y')
    try:
        P570_entity = wikidata.entity.Entity("P570", client)
        date_of_death = entity_dict[P570_entity].strftime('%d.%m.%Y')
    except KeyError:
        date_of_death = None
    
    return [date_of_birth, date_of_death]
```

```python
doplnena_data = {}
for key, value in wikid.items():
    doplnena_data[key] = [wikidates(value)[0], wikidates(value)[1]]
```
