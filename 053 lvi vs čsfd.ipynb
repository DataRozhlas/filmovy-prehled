```python
import requests
from bs4 import BeautifulSoup 
```

```python
def dopln_hodnoceni(x):
    response = requests.get(x, headers = {'User-agent': 'your bot 0.1'})
    soup = BeautifulSoup(response.content, 'html.parser')
    procenta = soup.find(class_="film-rating-average").text.strip().replace("%","")
    pocet = soup.find(class_="counter").text.strip().replace("(","").replace(")","").replace(" ","").replace("\xa0","")
    return [int(procenta), int(pocet)]
```

```python
dopln_hodnoceni("https://www.csfd.cz/film/710346-il-boemo/prehled/")
```

```python
doplnene_hodnoceni = []
for key, val in odkazy.items():
    skore = {}
    vysledek = dopln_hodnoceni(val)
    skore['Film'] = key
    skore['Hodnocení'] = vysledek[0]
    skore['Počet hodnocení'] = vysledek[1]
    print(skore)
    doplnene_hodnoceni.append(skore)
```
