```python
import os
from bs4 import BeautifulSoup
```

```python
slozka = "filmy"
```

```python
def seznam_souboru(slozka):
    seznam = []
    for file in os.listdir(slozka):
        seznam.append(file)
    return(seznam)
```

```python
def id_persony(url):
    if "cs/person/" in url:
        id_persony = url.split("/")[-2]
    else:
        id_persony = 0
    return(id_persony)
```

```python
def scrape_skladby(slozka, soubor):

    skladby = ""
    
    with open(os.path.join(slozka, soubor), "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")
            
    hudba = {}
    hudba["Film"] = f"""{soup.find("h1").text.strip()} ({soup.find('meta', {'property': 'og:url'})['content'].split("/")[-2]})"""
    
    credit = soup.find_all(class_="credit-item")
    
    for c in credit:
        
        if "Použitá hudba" in c.text:
            
            skladby = c.find("p").text.replace(" (",": ").split("),")
            
            skladby = [s.replace(")","").strip() for s in skladby]
            
            print(skladby)
                
    hudba["Skladby"] = skladby
        
    return(hudba)
```

```python
import pandas as pd
```

```python
kolik = None
```

```python
def frejm(slozka):
    pouzita_hudba = pd.DataFrame()
    total = len(seznam_souboru(slozka)[:kolik])
    count = 0
    for s in seznam_souboru(slozka)[:kolik]:
        count = count + 1
        print(f"{count} / {total} / {s}")
        line = pd.DataFrame.from_dict([scrape_skladby(slozka, s)])
        pouzita_hudba = pd.concat([pouzita_hudba, line])
    pouzita_hudba = pouzita_hudba.reset_index(drop = True)
    return(pouzita_hudba)
```

```python
p_hudba = frejm("filmy")
```

```python
p_hudba = p_hudba.replace(r'^\s*$', None, regex=True)
```
